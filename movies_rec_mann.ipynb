```python
import ast

def parse_genres_safely(genres_str):
    if isinstance(genres_str, str):
        try:
            # Try to evaluate if it's a string representation of a list of dicts
            genres_list = ast.literal_eval(genres_str)
            if isinstance(genres_list, list) and all(isinstance(g, dict) and 'name' in g for g in genres_list):
                return " ".join([g['name'] for g in genres_list])
        except (ValueError, SyntaxError):
            pass # If it's not a literal or already processed, return as is
    return genres_str

df['genres'] = df['genres'].apply(parse_genres_safely)
```

```python
# Check for genres that are still empty lists or not properly parsed
problematic_genres = df[df['genres'].apply(lambda x: x == '[]' or (isinstance(x, str) and x.strip().startswith('[')))]

if not problematic_genres.empty:
    print(f"Found {len(problematic_genres)} rows with potentially problematic genre formats:")
    print(problematic_genres['genres'].value_counts())
else:
    print("All genres appear to be processed correctly.")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python


nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python

def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #removed the punctuations
  text = re.sub(r'[^a-zA-Z\s]', '', text)

  words = text.split()
  #remove stopwords
  words = [word for word in words if word not in stop_words]

  #lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Avatar')
```

```python
# Simple notebook recommender example
import os
import requests
from IPython.display import display, Image

TMDB_API_KEY = os.getenv("TMDB_API_KEY", "8265bd1679663a7ea12ac168da84d2e8")
TMDB_IMG_BASE = "https://image.tmdb.org/t/p/w500"

def fetch_poster(movie_id):
    url = f"https://api.themoviedb.org/3/movie/{movie_id}"
    params = {"api_key": TMDB_API_KEY, "language": "en-US"}
    resp = requests.get(url, params=params, timeout=10)
    if resp.status_code != 200:
        return None
    data = resp.json()
    poster_path = data.get("poster_path")
    return f"{TMDB_IMG_BASE}{poster_path}" if poster_path else None

def recommend_notebook(title, n=5):
    if title not in indices:
        return [], []
    idx = indices[title]
    sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]
    titles = df['title'].iloc[similar_idx].tolist()
    posters = []
    for i in similar_idx:
        movie_id = int(df['id'].iloc[i])
        posters.append(fetch_poster(movie_id))
    return titles, posters

selected_movie = 'Avatar'  # change this to any movie title from df['title']
recommended_titles, recommended_posters = recommend_notebook(selected_movie, n=5)

print("Selected movie:", selected_movie)
for title, poster_url in zip(recommended_titles, recommended_posters):
    print("-", title)
    if poster_url:
        display(Image(url=poster_url, width=180))
    else:
        print("  No poster available")
```
